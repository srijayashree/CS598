```python
# create dataframe with 25,000 Records for easy processing

df2 = pd.read_csv("C:/Users/vrl2k/Desktop/MS/Spring 2023/Deep Learning/Project/MIMIC db/patient_records.csv")
```

```python
# randomly select 25,000 rows from NOTEEVENTS.csv (downloaed from MIMIC-iii)
df3 = df2.sample(n=25000, random_state=42)
```

```python
df3
```

```python
# drop unnecessary columns
df_clean = df3.drop(['CHARTTIME', 'STORETIME','ISERROR','CHARTDATE'], axis=1)
```

```python
# remove duplicate rows
df_dedup = df_clean.drop_duplicates()
```

```python
# replace all special characters with space in the 'text' column
df_dedup['TEXT'] = df_dedup['TEXT'].replace('[^a-zA-Z0-9]', ' ', regex=True)
```

```python
# replace numbers with space
df_dedup = df_dedup.replace('\d+', ' ', regex=True)
```

```python
# export the cleaned data to a .csv file
df_dedup.to_csv('C:/Users/vrl2k/Desktop/MS/Spring 2023/Deep Learning/Project/MIMIC db/patient_records_clean.csv', index=False)
```

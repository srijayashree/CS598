```python
import pandas as pd

# Read CSV file
df1 = pd.read_csv("C:/Users/vrl2k/Desktop/MS/Spring 2023/Deep Learning/Project/MIMIC db/patient_records_clean.csv")
```

```python
# Read DIAGNOSES_ICD.csv file and copy to df_subid
df_subid = pd.read_csv("C:/Users/vrl2k/Desktop/MS/Spring 2023/Deep Learning/Project/MIMIC db/DIAGNOSES_ICD.csv")
```

```python
# Read D_ICD_DIAGNOSES.csv file and copy to df_subid
df_icd = pd.read_csv("C:/Users/vrl2k/Desktop/MS/Spring 2023/Deep Learning/Project/MIMIC db/D_ICD_DIAGNOSES.csv")
```

```python
df_subid
```

```python
# Join patient_records with respective SUBJECT_ID to lookup ICD9_CODE
merged_df1 = pd.merge(df1, df_subid, on='SUBJECT_ID')
```

```python
merged_df1
```

```python
# drop unnecessary columns
merged_df1 = merged_df1.drop(['ROW_ID_y', 'HADM_ID_y'], axis=1)
```

```python
merged_df1
```

```python
# rename the columns
merged_df1 = merged_df1.rename(columns={'ROW_ID_x': 'ROW_ID', 'HADM_ID_x': 'HADM_ID'})
```

```python
merged_df1
```

```python
# rearrange the columns
merged_df1 = merged_df1.reindex(columns=['ROW_ID', 'SUBJECT_ID', 'HADM_ID', 'SEQ_NUM','ICD9_CODE','CATEGORY','DESCRIPTION','CGID','TEXT'])
```

```python
# Join patient_records with ICD9_CODE to add ICD short_title and long_title 
merged_df2 = pd.merge(merged_df1, df_icd, on='ICD9_CODE')
```

```python
merged_df2
```

```python
# drop unnecessary columns
merged_df2 = merged_df2.drop(['ROW_ID_y'], axis=1)
```

```python
merged_df2
```

```python
clean_df1=merged_df2
```

```python
#check for ICD-9 code for keywords 'Heart', 'Lung', 'Diabetes'
# Define the keywords to search for
keywords = ['heart', 'lung', 'diabetes']

# Create a new column with True if any of the keywords are found in the clinical text, False otherwise
clean_df1['condition_lookup'] = clean_df1['LONG_TITLE'].str.contains('|'.join(keywords), case=False)
```

```python
clean_df1
```

```python
filtered_df1 = clean_df1[clean_df1['condition_lookup'] == True]
```

```python
filtered_df1
```

```python
# remove duplicate rows
dedup_df1 = filtered_df1.drop_duplicates()
```

```python
dedup_df1
```

```python
# drop unnecessary columns
df1_drop = dedup_df1.drop(['ROW_ID_x','condition_lookup'], axis=1)
```

```python
df1_final = df1_drop
```

```python
# export the cleaned dataframe df1_final to a .csv file
df1_final.to_csv('C:/Users/vrl2k/Desktop/MS/Spring 2023/Deep Learning/Project/MIMIC db/patient_records_input.csv', index=False)
```
